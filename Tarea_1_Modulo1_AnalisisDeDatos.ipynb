```python
r,g,b=cv.split(imagen)
plt.figure(figsize=(18,9))
plt.subplot(311)
im=plt.imshow(b,cmap="Blues")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(312)
im=plt.imshow(r,cmap="Reds")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.subplot(313)
im=plt.imshow(g,cmap="Greens")
plt.colorbar(im,fraction=0.029, pad=0.04)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.hist(b.ravel(),bins=256,range=(0,256),color='b',label='blue')
plt.hist(g.ravel(),bins=256,range=(0,256),color='g',label='green')
plt.hist(r.ravel(),bins=256,range=(0,256),color='r',label='red')
plt.legend()
plt.show()
b1=b.ravel()
r1=r.ravel()
g1=g.ravel()
```

```python
print("la media es G acotada: ",np.mean(g1))
print("la media es R acotada: ",np.mean(r1))
print("la media es A acotada: ",np.mean(b1))
print("la mediana es G acotada: ",np.median(g1))
print("la mediana es R acotada: ",np.median(r1))
print("la mediana es A acotada: ",np.median(b1))
```

```python
#con Ses
print("la media es G acotada: ",np.mean(g1[(g1>180) & (g1<190)]))
print("la media es R acotada: ",np.mean(r1[(r1>250) & (r1<270)]))
print("la media es A acotada: ",np.mean(b1[(b1>155) & (b1<170)]))
print("la mediana es G acotada: ",np.median(g1))
print("la mediana es R acotada: ",np.median(r1))
print("la mediana es A acotada: ",np.median(b1))
```

```python
print("La Kurtosis es R",stats.kurtosis(r1))
print("La Kurtosis es B",stats.kurtosis(b1))
print("La Kurtosis es G",stats.kurtosis(g1))
print("Coeficiente de simetria es ", stats.skew(r1))
stats.mode(r1)
```
